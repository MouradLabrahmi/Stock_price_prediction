```
dataset_train = pd.read_csv('NSE-TATAGLOBAL.csv')
training_set = dataset_train.iloc[:, 1:2]
dataset_train.head()
```

```
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)
training_set_scaled
```

```
X_train = []
y_train = []
for i in range(60, 2035):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
```

```
X_train, y_train = np.array(X_train), np.array(y_train)
X_train.shape[1]
```

```
X_train
```

```
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```
regressor = Sequential()

regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 1))

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```
#NewDataSet = dataset_train.resample('D').mean()
TestData = dataset_train.tail(100)
TestData
```

```
real_stock_price = TestData.iloc[:, 1:2].values
```

```
dataset_total = pd.concat((dataset_train['Open'], TestData['Open']), axis = 0)
dataset_total
```

```
inputs = dataset_total[len(dataset_total) - len(TestData) - 60:].values
inputs
```

```
len(inputs)
```

```
inputs = inputs.reshape(-1,1)
print(len(inputs))
inputs
```

```
inputs = sc.transform(inputs)
print(len(inputs))
inputs
```

```
X_test = []
for i in range(60,160):
    X_test.append(inputs[i-60:i, 0])
len(X_test)
```

```
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price
```

```
len(X_test)
```

```
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```
len(predicted_stock_price)
```

```
plt.plot(real_stock_price, color = 'green', label = ' Real Stock Price')
plt.plot(predicted_stock_price, color = 'red', label = 'Predicted  Stock Price')
plt.title(' Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel(' Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```
